```python
df.drop(['CODE'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'OdName' : 'Nazione', 'AreaName' : 'Area', 'RegName' : 'Regione', 'DevName' : 'Sviluppo'}, inplace=True)
```

```python
df.head()
```

```python
df.set_index(['Nazione'], inplace=True)
```

```python
df.head()
```

```python
anni = list(map(str, range(1980, 2009)))
anni
```

```python
df.loc['Andorra', anni]
```

```python
df_ger_ita = df.loc[['Germany', 'Italy'], anni] 
df_ger_ita
```

```python
df_ger_ita = df_ger_ita.transpose()
df_ger_ita
```

```python
df_ger_ita.index = df_ger_ita.index.map(int) 
df_ger_ita.plot(kind='line', figsize=(12, 6))

plt.title('Immigrazione da Germania e Italia')
plt.ylabel('Numero migranti')
plt.xlabel('Anno')

plt.show()
```

```python
df['Total'] = df.sum(axis=1)
```

```python
df.head()
```

```python
inplace = True 
df.sort_values(by='Total', ascending=False, axis=0, inplace=True) 

df_top5 = df.head(5) 
df_top5 = df_top5[anni].transpose()

print(df_top5)
```

```python
df_top5.index = df_top5.index.map(int) 
df_top5.plot(kind='line', figsize=(18, 9)) 

plt.title('Immigrazione top 5 paesi') 
plt.ylabel('Numero migranti')
plt.xlabel('Anno') 

plt.show()
```
